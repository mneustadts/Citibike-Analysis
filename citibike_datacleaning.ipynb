```python
import pandas as pd
import numpy as np
import random
```

```python
aug_csv = "data/citibike_tripdata_Aug19.csv"

p= 0.1

aug_data = pd.read_csv(aug_csv,
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p)
aug_data.head()
```

```python
aug_data.count()
```

```python
sept_csv = "data/citibike_tripdata_Sept19.csv"

p= 0.1

sept_data = pd.read_csv(sept_csv,
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p)
sept_data.head()
```

```python
oct_csv = "data/citibike_tripdata_Oct19.csv"

p= 0.1

oct_data = pd.read_csv(oct_csv,
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p)
oct_data.head()
```

```python
nov_csv = "data/citibike_tripdata_Nov19.csv"

p= 0.1

nov_data = pd.read_csv(nov_csv,
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p)
nov_data.head()
```

```python
dec_csv = "data/citibike_tripdata_Dec19.csv"

p= 0.1

dec_data = pd.read_csv(dec_csv,
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p)
dec_data.head()
```

```python
july_csv = "data/citibike_tripdata_July19.csv"

p= 0.1

july_data = pd.read_csv(july_csv,
         header=0, 
         skiprows=lambda i: i>0 and random.random() > p)
july_data.head()
```

```python
july_data.to_csv("clean_data/July_data.csv")
aug_data.to_csv("clean_data/Aug_data.csv")
sept_data.to_csv("clean_data/Sept_data.csv")
oct_data.to_csv("clean_data/Oct_data.csv")
nov_data.to_csv("clean_data/Nov_data.csv")
dec_data.to_csv("clean_data/Dec_data.csv")
```

```python
six_month_data = pd.concat([july_data, aug_data, sept_data, oct_data, nov_data, dec_data], axis=0, join='inner')

six_month_data
```

```python
six_month_data.to_csv("clean_data/six_month_data.csv")
```
